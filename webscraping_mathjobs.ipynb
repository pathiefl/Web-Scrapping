# Math Jobs Web Scraping Project



```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time

# Step 1: Get all job links from the main jobs page
url = 'https://www.mathjobs.org/jobs?joblist-0-0------'
response = requests.get(url)
soup = bs(response.content)
div_by_institution_list = soup.find_all('div', class_="clr")

institution_link = []
for div_by_institution in div_by_institution_list:
  li_tags_list = div_by_institution.find_all('li')
  for li_tag in li_tags_list:
    a_tag = li_tag.find('a')
    institution_link.append('https://www.mathjobs.org/' + a_tag['href'])

list_of_urls = institution_link

# Step 2: Prepare empty lists to collect data
Name_list = []
ID_list = []
Title_list = []
Type_list = []
Location_list = []
Subject_list = []
Deadline_list = []
Link_list = []

# Step 3: Loop through all job links
for url in list_of_urls:

  response = requests.get(url)

  # For missing value
  Name = ''
  ID = ''
  Title = ''
  Type = ''
  Location = ''
  Subject = ''
  Deadline = ''
  Link = ''

  soup = bs(response.content)
  name = soup.title.text
  grid = soup.find('div', class_="grid2 nopd")

  if grid:
    divs_list = grid.find_all('div')
    for i in range(0,len(divs_list),2):
      key = divs_list[i].text
      if 'ID' in key:
        ID = divs_list[i+1].text
      if '[' in ID:  #this is what i changed to scrape only ID number
        ID = ID[ID.find('['):].strip()  #this is what i changed to scrape only ID number
      if 'Title' in key:
        Title = divs_list[i+1].text
      if 'Type' in key:
        Type = divs_list[i+1].text
      if 'Location' in key:
        Location = divs_list[i+1].text
      if '[map]' in Location:  #this is what i changed to get rid of [map]
        Location = Location.replace('[map]', '').strip()  #this is what i changed to get rid of [map]
      if 'Subject' in key:
        Subject = divs_list[i+1].text
      if 'Deadline' in key:
        Deadline = divs_list[i+1].text.split('(')[0]
      if 'Description' in key:
        a_tag = divs_list[i+1].find('a')
        if a_tag:
          Link = a_tag['href']

  #this is what i added to give NaN in empty spaces
  if Name == '': Name = 'NaN'
  if ID == '': ID = 'NaN'
  if Title == '': Title = 'NaN'
  if Type == '': Type = 'NaN'
  if Location == '': Location = 'NaN'
  if Subject == '': Subject = 'NaN'
  if Deadline == '': Deadline = 'NaN'
  if Link == '': Link = 'NaN'

  Name_list.append(name)
  ID_list.append(ID)
  Title_list.append(Title)
  Type_list.append(Type)
  Location_list.append(Location)
  Subject_list.append(Subject)
  Deadline_list.append(Deadline)
  Link_list.append(Link)

# Step 4: Create dataframe
data_dictionary = {
  'Instituion Name': Name_list,
  'Position ID': ID_list,
  'Title': Title_list,
  'Type': Type_list,
  'Location': Location_list,
  'Subject': Subject_list,
  'Deadline': Deadline_list,
  'Link': Link_list
}

df = pd.DataFrame(data_dictionary)
df
```
